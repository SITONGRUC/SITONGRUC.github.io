```python
import plotly.graph_objects as go
import numpy as np
import plotly.express as px
import pandas as pd

```

```python
data_path = './data/'
item1A_bert = pd.read_excel(data_path+'1A_BERT_Analysis_sum_scaled.xlsx')
```

```python
# Environment
environment = [
    "Physical_Impacts_Of_Climate_Change",
    "Waste_And_Hazardous_Materials_Management",
    "Water_And_Wastewater_Management",
    "Air_Quality",
    "Ecological_Impacts",
    "Energy_Management",
    "GHG_Emissions"
]

# Social
social = [
    "Business_Ethics",
    "Access_And_Affordability",
    "Customer_Welfare",
    "Employee_Engagement_Inclusion_And_Diversity",
    "Employee_Health_And_Safety",
    "Human_Rights_And_Community_Relations",
    "Labor_Practices",
    "Product_Quality_And_Safety",
    "Product_Design_And_Lifecycle_Management",
    "Selling_Practices_And_Product_Labeling",
    "Supply_Chain_Management",
    "Customer_Privacy"
]

# Governance
governance = [
    "Data_Security",
    "Business_Model_Resilience",
    "Competitive_Behavior",
    "Critical_Incident_Risk_Management",
    "Director_Removal",
    "Management_Of_Legal_And_Regulatory_Framework",
    "Systemic_Risk_Management"
]

item1A_bert['environment']= item1A_bert[environment].T.sum()
item1A_bert['social']= item1A_bert[social].T.sum()
item1A_bert['governance']= item1A_bert[governance].T.sum()
```

```python
ESG_fig = go.Figure()
for i in ['environment','social','governance']:
     ESG_fig.add_trace(go.Scatter(x=item1A_bert.groupby('year')[i].mean().index,
                              y = item1A_bert.groupby('year')[i].mean().values,
                         mode='lines',
                         name=i))
ESG_fig.update_layout(
    plot_bgcolor='#f9f9f9',  # Set the background color for the plot area
    paper_bgcolor='#f9f9f9',  # Set the background color for the entire figure
    font=dict(color='#333'),   # Set the font color
    legend=dict(
        orientation="h",  # Horizontal orientation for the legend
        y=-0.2,           # Position legend below the plot
        x=0.5,            # Center the legend horizontally
        xanchor='center',
        yanchor='top'
    ),
)
ESG_fig.update_layout(title='Average ESG disclosure score for US firms',
                   xaxis_title='Year',
                   yaxis_title='Score')


environment_fig = go.Figure()
for i in environment :
     environment_fig.add_trace(go.Scatter(x=item1A_bert.groupby('year')[i].mean().index,
                              y = item1A_bert.groupby('year')[i].mean().values,
                         mode='lines',
                         name=i))
environment_fig.update_layout(
    plot_bgcolor='#f9f9f9',  # Set the background color for the plot area
    paper_bgcolor='#f9f9f9',  # Set the background color for the entire figure
    font=dict(color='#333'),   # Set the font color
    legend=dict(
        orientation="h",  # Horizontal orientation for the legend
        y=-0.4,           # Position legend below the plot
        x=0.5,            # Center the legend horizontally
        xanchor='center',
        yanchor='top'
    ),
    # width=1200,  # Set the desired width of the figure
    height=700 
)
environment_fig.update_layout(title='Average environment disclosure score for US firms',
                   xaxis_title='Year',
                   yaxis_title='Score')

social_fig = go.Figure()
for i in social :
     social_fig.add_trace(go.Scatter(x=item1A_bert.groupby('year')[i].mean().index,
                              y = item1A_bert.groupby('year')[i].mean().values,
                         mode='lines',
                         name=i))
social_fig.update_layout(
    plot_bgcolor='#f9f9f9',  # Set the background color for the plot area
    paper_bgcolor='#f9f9f9',  # Set the background color for the entire figure
    font=dict(color='#333'),   # Set the font color
    legend=dict(
        orientation="h",  # Horizontal orientation for the legend
        y=-0.4,           # Position legend below the plot
        x=0.5,            # Center the legend horizontally
        xanchor='center',
        yanchor='top'
    ),
    # width=1200,  # Set the desired width of the figure
    height=700 
)
social_fig.update_layout(title='Average social disclosure score for US firms',
                   xaxis_title='Year',
                   yaxis_title='Score')
governance_fig = go.Figure()
for i in governance :
     governance_fig.add_trace(go.Scatter(x=item1A_bert.groupby('year')[i].mean().index,
                              y = item1A_bert.groupby('year')[i].mean().values,
                         mode='lines',
                         name=i))
governance_fig.update_layout(
    plot_bgcolor='#f9f9f9',  # Set the background color for the plot area
    paper_bgcolor='#f9f9f9',  # Set the background color for the entire figure
    font=dict(color='#333'),   # Set the font color
    legend=dict(
        orientation="h",  # Horizontal orientation for the legend
        y=-0.4,           # Position legend below the plot
        x=0.5,            # Center the legend horizontally
        xanchor='center',
        yanchor='top'
    ),
    # width=1200,  # Set the desired width of the figure
    height=700 
)
governance_fig.update_layout(title='Average governance disclosure score for US firms',
                   xaxis_title='Year',
                   yaxis_title='Score')

import plotly.express as px
from jinja2 import Template




output_html_path="social.html"

input_template_path = "file copy.html"

plotly_jinja_data = {
    "fig1": ESG_fig.to_html(full_html=False),
    "fig2":environment_fig.to_html(full_html=False),
        "fig3":social_fig.to_html(full_html=False),
                "fig4":governance_fig.to_html(full_html=False)
        
    
}

# plotly_jinja_data = {"fig2":fig.to_html(full_html=False)}
#consider also defining the include_plotlyjs parameter to point to an external Plotly.js as described above

with open(output_html_path, "w", encoding="utf-8") as output_file:
    with open(input_template_path) as template_file:
        j2_template = Template(template_file.read())
        output_file.write(j2_template.render(plotly_jinja_data))

        
```

```python
data_total = pd.DataFrame()
for i in range(2012,2024):
    data = pd.read_csv(data_path+str(i)+'_0.9stencen_count.csv')
    data_total = pd.concat([data_total,data])
data_total['year'] = data_total['index'].apply(lambda x:x[:4])
data_total['environment']= data_total[environment].T.sum()
for i in environment:
    data_total[i] = data_total[i].apply(lambda x:1 if x >5 else 0)
ESG_fig = go.Figure()
for i in environment:
     data_for_draw = data_total.groupby('year')[i].sum()*100/data_total.groupby('year')[i].count()
     ESG_fig.add_trace(go.Scatter(x=data_for_draw.index,
                              y = data_for_draw.values,
                         mode='lines',
                         name=i))
ESG_fig.update_layout(
    plot_bgcolor='#f9f9f9',  # Set the background color for the plot area
    paper_bgcolor='#f9f9f9',  # Set the background color for the entire figure
    font=dict(color='#333'),   # Set the font color
    legend=dict(
        orientation="h",  # Horizontal orientation for the legend
        y=-0.3,           # Position legend below the plot
        x=0.5,            # Center the legend horizontally
        xanchor='center',
        yanchor='top'
    ),
    height=700 ,
)
ESG_fig.update_layout(title='Percentage that US firms choose to disclosure environment',
                   xaxis_title='Year',
                   yaxis_title='Score')
```

```python
output_html_path="percentage.html"

input_template_path = "file copy.html"

plotly_jinja_data = {
    "fig1": ESG_fig.to_html(full_html=False),  
}

# plotly_jinja_data = {"fig2":fig.to_html(full_html=False)}
#consider also defining the include_plotlyjs parameter to point to an external Plotly.js as described above

with open(output_html_path, "w", encoding="utf-8") as output_file:
    with open(input_template_path) as template_file:
        j2_template = Template(template_file.read())
        output_file.write(j2_template.render(plotly_jinja_data))
```
